```python

# Celda 2: Cargar el dataset desde GitHub

url_csv = "https://raw.githubusercontent.com/dtoralg/IE_Calidad_ML/refs/heads/main/Data/control_calidad_piezas_metalicas.csv"
df = pd.read_csv(url_csv)

# Mostrar las primeras filas del dataset
df.head()
```

```python
# Celda 3: Inspecci贸n de la estructura del dataset

print("N煤mero de filas y columnas:", df.shape)
print("Tipos de datos:")
print(df.dtypes)
```

```python
# Celda 4: Valores 煤nicos por columna

print("Valores 煤nicos en cada columna:")
print(df.nunique())
```

```python
# Celda 5: Estad铆sticas descriptivas b谩sicas

df.describe(include="all")
```

```python
# Celda 6: Identificaci贸n de valores nulos

print("Valores nulos por columna:")
print(df.isnull().sum())

# Visualizaci贸n de proporci贸n de valores nulos
plt.figure(figsize=(8, 4))
sns.heatmap(df.isnull(), cmap="viridis", cbar=False, yticklabels=False)
plt.title("Mapa de valores nulos en el dataset")
plt.show()
```

```python

# Celda 7: Visualizaci贸n de datos

# Histograma de la variable Temperatura
plt.figure(figsize=(8,5))
sns.histplot(df["Temperatura"], bins=30, kde=True)
plt.title("Distribuci贸n de la Temperatura de las Piezas")
plt.xlabel("Temperatura (掳C)")
plt.ylabel("Frecuencia")
plt.show()

# Gr谩fico de barras para la variable Defecto
plt.figure(figsize=(8,5))
sns.countplot(x="Defecto", data=df, order=df["Defecto"].value_counts().index)
plt.title("Frecuencia de Defectos en Piezas Met谩licas")
plt.xlabel("Defecto")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.show()
```
